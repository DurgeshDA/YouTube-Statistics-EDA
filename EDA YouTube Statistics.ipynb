```python
pd.set_option("display.float_format", "{:.3f}".format)
```

```python
category_subscribers_mean = df.groupby("category")["subscribers"].mean()
category_subscribers_mean
```

```python
category_subscribers_mean.sort_values(ascending = False, inplace = True)
```

```python
category_subscribers_mean.head(7).plot(kind = 'bar')
plt.title("Mean Subscribers by Category")
plt.xlabel("Category")
plt.ylabel("Mean Subscribers")
plt.xticks(rotation = 90)
plt.show()
```

```python
channel_type_counts = df["channel_type"].value_counts()
channel_type_counts
```

```python
channel_type_counts.head(7).plot(kind = 'pie', autopct = "%1.1f%%")
plt.title("Channel Type Distribution")
plt.show()
```

Relation between subscribers and video views

```python
plt.scatter(df["subscribers"], df["video views"])
plt.title("Scatter Plot of Subscribers vs Video Views")
plt.xlabel("Subscribers")
plt.ylabel("Video Views")
plt.show()
```

Observing Country Wise Subscribers Distribution

```python
country_subscribers_total = df.groupby("Country")["subscribers"].sum().sort_values(ascending = False)
```

```python
country_subscribers_total
```

```python
country_subscribers_total.head(10).plot(kind = "bar")
plt.title("Total Subscribers by Country")
plt.xlabel("Country")
plt.ylabel("Total Subscribers")
plt.xticks(rotation = 90)
plt.show()
```

Observating Data on the Basis of Created Year of Channel

```python
df.groupby("created_year")["subscribers"].mean().sort_values(ascending = True)
```

```python
pd.set_option("display.float_format", "{:.0f}".format)
```

```python
df.groupby("channel_type")["video_views_rank"].median().sort_values(ascending = False)
```

Finding Maximum Subscribers for last 30 Days 

```python
df.groupby("Country")["subscribers_for_last_30_days"].max().sort_values(ascending = False)
```

Highest Monthly Earning by Channel Type

```python
sns.boxplot(x = "channel_type", y = "highest_monthly_earnings", data = df)
plt.title("Distribution of Highest Monthly Earning by Channel Type")
plt.xticks(rotation = 90)
plt.show()
```

```python
df.groupby("channel_type")["Unemployment rate"].mean().sort_values(ascending = False).head(5).plot(kind = "bar")
plt.title("Average Unemployment Rate by Channel Type")
plt.xlabel("Channel Type")
plt.ylabel("Average Unemployment Rate")
plt.xticks(rotation = 90)
plt.show()
```

Average Subscribers by created years and channel type

```python
df.groupby(["created_year", "channel_type"])["subscribers"].mean().unstack().tail().plot(kind = "bar", width = 1)
plt.title("Average Subscribers by Created Year and Channel Type")
plt.xlabel("Created Year")
plt.ylabel("Average Subscribers")
plt.xticks(rotation = 45)
plt.legend(title = "Channel Type", bbox_to_anchor = (1.05,1), loc = "upper left")
plt.show()
```

Highest Monthly Earnings by Category

```python
category_highest_monthly_earnings_sum = df.groupby("category")["highest_monthly_earnings"].sum()
```
